```python
import pandas as pd
import seaborn as sns
import sklearn
```

```python
df=pd.read_csv("/content/drive/MyDrive/AI/ML/Advertisment_sales2/advertising.csv")
df.head(6)
```

```python
df.columns
```

```python
x = df[['TV', 'Radio', 'Newspaper']]
```

```python
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=15)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
LR.fit(x_train.values,y_train)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
print("The LR model is: Y = ",LR.intercept_, "+", LR.coef_[0], "TV + ", LR.coef_[1], "radio + ", LR.coef_[2], "newspaper")
```

```python
advertisment = [[120, 50, 60]]

pred = LR.predict(advertisment)
print("prediction= ", pred[0])
```

```python
y_pred = LR.predict(x_test.values)
y_pred
```

```python
sns.regplot(x = y_test, y = y_pred, color='green')
```

```python
from sklearn import metrics
import numpy as np

mae = metrics.mean_absolute_error(y_test,y_pred)
mse = metrics.mean_squared_error(y_test,y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
print('mae =',round(mae,2))
print('mse =',round(mse,2))
print('rmse =',round(rmse,2))
```
