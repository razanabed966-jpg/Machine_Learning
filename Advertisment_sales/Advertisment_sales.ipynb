```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv("/content/drive/MyDrive/AI/ML/Advertisment_sales/advertising.csv")
df.head(6)
```

```python
df.isnull().sum()
```

```python
pd.DataFrame(df.isnull().sum(), columns =["Count of Null Values"]).T
```

```python
X = df[['TV']]
```

```python
Y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train.values, Y_train)
```

```python
b0 = round(model.intercept_,2)
b1 = round(model.coef_[0],2)

print('Intercept (b0):', b0)
print('Coefficient (b1):', b1)
```

```python
Tv_advertisment = [[120]]

pred = model.predict(Tv_advertisment)
print("prediction= ", pred[0])
```

```python
Tv_advertisment
```

```python
b0 = model.intercept_
b1 = model.coef_[0]
a = Tv_advertisment[0][0]
pred2 = b0 + b1 * a
print("prediction = ", pred2)
```

```python
Y_pred = model.predict(X_test.values)
```

```python
Y_pred
```

```python
diff = pd.DataFrame({'Actual':Y_test,'Predicted':Y_pred})
diff.head(10)
```

```python
from sklearn import metrics
import numpy as np

mae = metrics.mean_absolute_error(Y_test,Y_pred)
mse = metrics.mean_squared_error(Y_test,Y_pred)
rmse = np.sqrt(metrics.mean_squared_error(Y_test,Y_pred))
```

```python
print('mae =', round(mae, 2))
print('mse =', round(mse, 2))
print('rmse =', round(rmse, 2))
```
